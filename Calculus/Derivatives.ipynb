## Definition of the derivative  
  
The  derivative is the answer to the question "how do we find the slope of a function at a particular point. To find the slope of a line, we use the formula: $ m = \frac{y_2 - y_1}{x_2 - x_1} $. The derivative of is what happens when the distance between two points approches 0. When the points are close enough together, the points represent a line "tangent" to the graph (basically a line that shows the slope of the graph at a certain point).

```python
import numpy as np
import matplotlib.pyplot as plt
# Define parabola
def f(x): 
    return x**2

# Define parabola derivative
def slope(x): 
    return 2*x

# Define x data range for parabola
x = np.linspace(-5,5,100)

# Choose point to plot tangent line
x1 = -3
y1 = f(x1)

# Define tangent line
# y = m*(x - x1) + y1
def line(x, x1, y1):
    return slope(x1)*(x - x1) + y1

# Define x data range for tangent line
xrange = np.linspace(x1-1, x1+1, 10)

# Plot the figure
plt.figure()
plt.plot(x, f(x))
plt.scatter(x1, y1, color='C1', s=50)
plt.plot(xrange, line(xrange, x1, y1), 'C1--', linewidth = 2)
```

Becasue of this, we can say that the "definition" of a derivative is :  
$$
f' (x) = \lim\limits_{h \to \ 0} \frac{f(x + h) - f(x)}{h}
$$
  
Remember that the formula is techincally $\frac{f(x + h) - f(x)}{(x + h) - x}$ (see slope of line formula above). The x's cancel leaving only the h behind.  
  
Example: Find the derivitive of $ f(x) = x^2 - 5x + 6 $  
  
$ f(x + h) = (x + h)^2 -5(x + h) + 6 $ -Substitute  

$ f'(x) = \lim\limits_{h \to \ 0} \frac{(x^2 + 2xh + h^2 -5x -5h + 6) - (x^2 - 5x + 6)}{h} \to  \frac{x^2 + 2xh + h^2 -5x -5h + 6 - x^2 + 5x - 6}{h} $ - Expand and distribute negitive  

$ f'(x) = \lim\limits_{h \to \ 0} \frac{\cancel{x^2} + 2xh + h^2 - \cancel{5x} -5h + \cancel{6} - \cancel{x^2} + \cancel{5x} - \cancel{6}}{h} \to \frac{h^2 + 2xh - 5h}{h} \to \frac{\cancel{h}(h + 2x - 
5)}{\cancel{h}}$ -Combine and cancel terms  
  
$ f'(x) = \lim\limits_{h \to \ 0} h + 2x - 5 \to 0 + 2x - 5 $ -Substitute 0 for h  
  
$ f'(x) = 2x - 5 $ 

## Derivative Rules  
  
### Basic Rules Quick reference table  
  
\begin{array}{|c|c|c|}
\hline \textbf{Property}  &        \textbf{Explaination}                                       & \textbf{Formula} \\\hline
\text{Power Rule}  &  \text{Calcualte the derivative of power functions} & \frac{d}{dx}x^x  = nx^{n - 1}\\\hline

\text{Constant rule} &  \text{Calcualte the derivative of any constant function} & \frac{d}{dx}k = 0 \\\hline

\text{Constant multiple rule}  &  \text{Calcualte the derivative of a constant multiplied by a function } & \frac{d}{dx}[k * f(x)] = k * \frac{d}{dx}f(x) \\\hline

\text{Sum/Difference rule}  &  \text{Calcualte the derivative of a sum/difference of functions} & \frac{d}{dx}[f(x) \pm g(x)] = \frac{d}{dx}f(x) \pm \frac{d}{dx}g(x) \\\hline

\text{Product rule}  &  \text{Calcualte the derivative of a product of two functions} &  \frac{d}{dx}[f(x)g(x)] = f(x)g'(x) + f'(x)g(x) \\\hline

\text{Quotient rule}  &  \text{Calcualte the derivative of a quotient of two functions} &  \frac{d}{dx}(\frac{f(x)}{g(x)}) = \frac{f'(x)g(x) - f(x)g'(x)}{[g(x)]^2}\\\hline  

\text{Chain rule}  &  \text{Calcualte the derivative of nested functions} &  \frac{d}{dx}g[f(x)] = g'[f(x)]f'(x)    \\\hline 

\end{array}

### Power Rule: fractional powers  

A power function with a negative exponent $ x^{-a} $ can be turned positive by moving the power function to the denominator. IE $ x^{-a} = \frac{1}{x^a} $ OR $ \frac{1}{x^{-a}} = x^a $
    
You can use the power rule to differentiate fractional powers. Example: Find the derivatice of the fuction $ y = x^{\frac{3}{2}} + 6x^{-\frac{1}{4}}  $  
  
$ y' = (\frac{3}{2}) x^{ \frac{3}{2} - 1 } + 6(-\frac{1}{4})x^{ -\frac{1}{4} - 1 } $  
  
$ y' = (\frac{3}{2}) x^{ \frac{3}{2} - \frac{2}{2} } + 6(-\frac{1}{4})x^{ -\frac{1}{4} - \frac{4}{4} } $  
  
$ y' = \frac{3}{2} x^{\frac{1}{2}} -\frac{3}{2}x^{ -\frac{5}{4}} $ -Can rewrite so all exponets are positive, but not nessassary. See next line.
  
$ y' = \frac{3}{2} x^{\frac{1}{2}} -\frac{3}{2 x^{ \frac{5}{4}} } $
  
Roots can also be broken up as fractional exponets to help find the derivative. A few examples are:  
$ x^{\frac{5}{4}} = \sqrt[4]{x^5} $  
$ x^{\frac{2}{3}} = \sqrt[3]{x^2} $  
$ x^{\frac{1}{2}} = \sqrt{x} $  

### Product Rule: three or more functions  
  
For three or more functions, the product rule is just the derivative of one of the functions from the original equation, multiplied by the rest of functions from the original equation. So for and equation that is the product of four functions: $ y = ABCD $ the derivative will include each possible combination :  
$$
y' = A'BCD + AB'CD + ABC'D + ABCD'
$$  
  
Example: Find the derivative of $ y = (4x^6)(-2x)(-x^3) $
  
$ y' = (24x^5)(-2x)(-x^3) + (4x^6)(-2)(-x^3) + (4x^6)(-2x)(-3x^2) $  -Product rule
  
$ y' = 48x^9 + 8x^9 + 24x^9 $ -Factor
  
$ y' = 80x^9 $ -Simplyfy

### Quotient rule expanded  
  
The quotient rule says that the derivative of a quotient is given by:
$$
\frac{d}{dx} (\frac{f(x)}{g(x)}) = \frac{  f'(x)g(x) - f(x)g'(x)   }{[g(x)]^2}
$$  
  
This is similar to other rules were the derivative is a combination of the f(x) and g(x) function and derivatives.  
  
Example: Use the quotient rule to find the derivative of $ y = \frac{x^4}{6x^2} $  
  
$ y' = \frac{ (4x^3)(6x^2) - (x^4)(12x)    }{ (6x^2)^2 } $ -Quotient rule  
  
$ y' = \frac{ 24x^5 - 12x^5    }{ 36x^4 } $ -Factor  
  
$ y' = \frac{  12x^5 }{ 36x^4 } \to \frac{  x }{ 3 }  \to \frac{  1 }{ 3 }x$ -simplyfiy

#### Quotient rule : Reciprocal Rule  
  
When the numerator is a constant, we can use a simpler version of the Reciprocal Rule:  
$$
y' = \frac{  0g(x) - ag'(x) }{ [g(x)]^2 }  \to  \frac{ - ag'(x) }{ [g(x)]^2 }
$$  
  
This is not technially a "rule" because it is a modified version of the quotient rule, but can help save time. 

### Chain rule expanded  
  
The chain rule allows for finding the derivative of nested functions. It can be used in combination with the other rules and trig functions to find deriviatives.  
  
Example: Find the deriviative of $ y = sin^3(x^2) $  
  
$ y' = 3[sin(x^2)]^2 cos(x^2) $ -Apply chain rule and power rule  
  
$ y' = 3[sin(x^2)]^2 cos(x^2) 2x \to 6x sin^2(x^2)cos(x^2) $ -Apply chain rule again and simplyfy

## Logarithmic differentiation
  
### Laws of Logs/Natural Logs Reference Table  
  
\begin{array}{ c c }  
\hline \textbf{Laws of Logs}  &        \textbf{Laws of Natural Logs}     \\\hline
\log_a a^x = x  &  \ln[e^x] = x    \\\hline

a^{\log_a x} = x  &  e^{\ln[x]} = x    \\\hline

\log_a x^r = r \log_a x  &  \ln[x^a] = a \ln x   \\\hline

\log_a [xy] = \log_a x + \log_a y  &  \ln [xy] = \ln x + \ln y  \\\hline

\log_a \frac{x}{y} = \log_a x - \log_a y   &    \ln \frac{x}{y} = \ln x - \ln y  \\\hline


\end{array}  
  
Some problems are easiest to solve using logarithmic differentiation. Logarithmic differentiation is a problem-solving method in which we start by applying the natural log function to both sides of the equation. We use logarithmic differentiation when it’s easier to differentiate the logarithm of a function than the function itself.  
  
Example: Use logarithmic differentaiation to find the derivative $ y = 3^{5x} $  
  
$ \ln y = \ln[3^{5x}]  $ -Apply Natural log to both sides  
  
$ \ln y = 5x \ln[3] $ -Apply Natural log rule: $ \ln[x^a] = a \ln x $  
  
$ \frac{1}{y}y' = 5x * 0  + \ln[3] * 5 = 5\ln[3]  $ -Differentiate both sides, using the product rule on the left. Simplyfy.  
  
$ y * \frac{1}{y}y' = (y)(5\ln[3]) $ -Mutliply by y on both sides

## Tangent Lines  
  
A tangent line is a line that touches the graph at exactly one point (see graph above). The equation for a tangent line is as follows:  
$$
y = f(a) + f'(a)(x - a)
$$


Remember that the equation of a line in point slope form $ y - y_1 = m(x - x_1) $ In the equation above:   
- a  = a x-value point of the orginal function ( one that we are trying to find the tanget line for). This is $x_1$ in the point slope form  
- f(a) = the value of the funtion at the point a. This is $y_1$ in point slope form  
- f'(a) = The slope of the tanget line (the derivative) at a. This is m in point slope form

Example: Find the equation of the tangent line of function $g(x) = 3x^2 - 6x + 1$  

$ 3(1)^2 - 6(1) + 1 = -2 $ - Find funtion solution for x = 1  

$ g'(x) = 6x - 6 $ -Find g'(x)  
$ g'(1) = 6(1) - 6  =  0 $ -Find g'(x) at x = 1 tp find slope  
   
$ y = (-2) + (0)(x - 1) = -2 $ -Substitue back into formula: $ y = f(a) + f'(a)(x - a) $ and simpliyfy  
  
  
### Parallel Tangent Lines  
  
You can use the equation of a tangent line to find the relationships between multiple tangent lines of the same function.  
  
Example: Find the value of $a$ such that the tangent lines of $f(x)$ at x = a and x = a + 1  
  
$f(x) = x^3 +2x^2 - 3x + 1 $  
  
$ f'(x) = 3x^2 + 4x - 3 $ -Find the derivative of f(x)  
  
$ f'(a) = 3a^2 +4a - 3 $  
$ f'(a + 1) = 3(a + 1)^2 + 4(a + 1) - 3 $  -substitue x  for both a and (a + 1)  
  
$ 3a^2 +4a - 3 = 3(a + 1)^2 + 4(a + 1) - 3 $ -Set derivatives equal to each other  
  
$ 3a^2 + 4a - 3 = 3a^2 + 6a + 3 + 4a + 4 - 3  $  
$ 4a - 3 = 6a + 3 + 4a + 4 - 3 $  
$ 4a - 3 = 10a + 4 $  
$ -3 = 6a + 4 $  
$ -7 = 6a $
$ a = -\frac{7}{6} $ -Collect like terms and solve for a  
  
$ a + 1 = -\frac{7}{6} + 1 $  
$ a + 1 = -\frac{7}{6} + \frac{6}{6} $ 
$ a + 1 = -\frac{1}{6} $ -find the value of a + 1

Once the a values have been found, you can plug the $a$ values ($-\frac{7}{6},-\frac{1}{6}$) into the f(x) and f'(x) to find the equations of the tangent lines using the tangent line formula above.
